```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def count_of_ones_fractional(number:'float'):
    count = 0
    while number:
        count += int(number * 2)
        number = (number * 2) % 1
    return count
```

```python
def count_of_ones(number:'int'):
    count = 0
    while number:
        count += number % 2
        number //=2
    return count
```

```python
N = 15
```

```python
x = np.array([ k / 2 ** N for k in range(0, 2 ** N)])
x1 = np.array([ k  for k in range(0, 2 ** N)])
y = np.array([count_of_ones_fractional(k) for k in x])
y1 = np.array([count_of_ones(k) for k in x1])
```

```python
fig, ax = plt.subplots(figsize = (100,20))
plt.plot(x, y)
```

```python
fig, ax = plt.subplots(figsize = (100,20))
plt.plot(x1, y1)
```
