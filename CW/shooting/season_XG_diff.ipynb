```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tottenham prem xg.csv')
```

```python
df.head()
```

```python
xG = df['xG']
xGA = df['xGA'] 
```

```python
xG = xG[~np.isnan(xG)]
xGA = xGA[~np.isnan(xGA)]
match = np.arange(1,34,1)
xG.to_numpy()
xGA.to_numpy()
```

```python
plt.figure(figsize = (30,10))
plt.plot(match, xG, color = 'black')
plt.plot(match, xGA,  color = 'red')
```
